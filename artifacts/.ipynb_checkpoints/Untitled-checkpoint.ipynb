```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
X_train = np.random.rand(1000, 10, 3)  # 1000 samples, each with a sequence length of 10 and 3 features
y_train = np.random.randint(0, 2, size=(1000,))  # Binary labels for classification

X_test = np.random.rand(200, 10, 3)  # 200 samples for testing
y_test = np.random.randint(0, 2, size=(200,))
```

```python
model = Sequential([
    LSTM(50, input_shape=(10, 3)),  # LSTM layer with 50 units
    Dense(10, activation='sigmoid')  # Output layer for binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```
